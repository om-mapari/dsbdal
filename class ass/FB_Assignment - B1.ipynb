```python
# Dropping records containing null values
data = data.dropna(axis = 0)
data.shape
```

### Subsets of dataset

```python
# Data subsets
subset1 = data.loc[:, ["Type","Category","Page total likes"]]
subset2 = data.loc[:,["comment", "like", "share"]]
subset2.columns
```

### Merge datasets

```python
mergedData = pd.concat([subset1,subset2], ignore_index=True, axis=1)
mergedData
```

### Sorting data based on "Lifetime Engaged Users" and "Lifetime Post Consumptions" in descending order

```python
# Sort data
sortedData = data.sort_values(by=["Lifetime Engaged Users", "Lifetime Post Consumptions"], ascending=[False,False])
sortedData
```

### Transpose of the given dataset

```python
transposedData = data.transpose()
transposedData
```
